### Library and Modles

```python
from bs4 import BeautifulSoup
import requests
import csv
import time
import sys
from bs4 import BeautifulSoup
from selenium import webdriver
```

## Scraping **Trading Codes** to Generate the urls of different companies

```python
source = requests.get("https://www.dsebd.org/company_listing.php").text
soup = BeautifulSoup(source, "lxml")
Trading_Name = []
for trading_name in soup.find_all('a', class_='abhead'):
    Trading_Name.append(trading_name['href'].split("=")[1])
print(Trading_Name)
```

## Automatically Generating the urls of the companies

```python
Company_URL =[]
for name in Trading_Name:
    company_url = 'https://www.dsebd.org/displayCompany.php?name='+name
    print(company_url)
    Company_URL.append(company_url)
```

## Parsing from individual company pages

```python
# Chrome Driver setup using Selenium for Google Colab, which It will help to parse dynammic data

sys.path.insert(0,'/usr/lib/chromium-browser/chromedriver')
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
wd = webdriver.Chrome('chromedriver',options=chrome_options)
```

### Creating Header of the Dataframe and Parsing Data

#### Header

```python
url = 'https://www.dsebd.org/displayCompany.php?name=1JANATAMF'
wd.get(url)
html = wd.page_source
soup2 = BeautifulSoup(html,'lxml')

HEADER = ["Trading Code","Company Name"]

# [0] Market Info Table, [1] Basic Infor Table, [11] Company Address Table
""" MARKET INFORMATION"""
# Market Information Table
MI_table = soup2.find('table', class_="table table-bordered background-white", id="company")

# Headers/Name of data in to Market Information Table
# MI_headers = []  
for header in MI_table.find_all('th'):
    # print(header.text.strip())
    HEADER.append(header.text.strip())
HEADER.insert(6+2, "Change(%)") # inserting change(%) again because the html file does't contain data header


""" BASIC INFORMATION"""
BI_table = soup2.find_all('table', class_="table table-bordered background-white", id="company")[1]

# BI_headers = []
for header in BI_table.find_all('th')[0:8]:
    # print(header.text.strip())
    HEADER.append(header.text.strip())


""" ADDRESS OF THE COMPANY"""
Ad_table = soup2.find_all('table', class_="table table-bordered background-white", id="company")[11]

# Ad_headers = []
for i, td in enumerate(Ad_table.find_all('td')):
    data = td.text.strip()
    # print(i, data)
    if i%2==0:
        HEADER.append(data)

# HEADER.append(MI_headers)
# HEADER.append(BI_headers)
# HEADER.append(Ad_headers)

HEADER
```

```python
COMPANY_DATA_2D = [] #
# Scraping Dynamic Data
for i, name in enumerate(Trading_Name):
    url = 'https://www.dsebd.org/displayCompany.php?name='+name
    COMPANY_DATA = []
    COMPANY_DATA.append(name)

    print("\n\n")
    print(i, url)
    wd.get(url)
    html = wd.page_source
    soup2 = BeautifulSoup(html,'lxml')

    
    """ COMPANY NAME"""
    company_name = (soup2.find('h2', class_= "BodyHead topBodyHead").text.split(':')[1]).strip()
    COMPANY_DATA.append(company_name)
    print(company_name.strip())

    # [0] Market Info Table, [1] Basic Infor Table, [11] Company Address Table
    
    """ MARKET INFORMATION"""
    # Market Information Table
    MI_table = soup2.find('table', class_="table table-bordered background-white", id="company")
    print("MI Table -- ok")
   
   #Market Information Data
    MI_data = []
    for td in MI_table.find_all('td'):
        data = td.text.strip()
        COMPANY_DATA.append(data)
    print("MI Data -- ok")



    """ BASIC INFORMATION"""
    BI_table = soup2.find_all('table', class_="table table-bordered background-white", id="company")[1]
    print("BI Table -- ok")

    BI_data = []
    for td in BI_table.find_all('td')[0:8]:
        data = td.text.strip()
        # print(data)
        COMPANY_DATA.append(data)
    print("BI Data -- ok")


    """ ADDRESS OF THE COMPANY"""
    Ad_table = soup2.find_all('table', class_="table table-bordered background-white", id="company")[11]
    print("Address Table -- ok")

    # Ad_headers = []
    Ad_data = []
    for i, td in enumerate(Ad_table.find_all('td')):
        data = td.text.strip()
        # print(i, data)
        # if i%2==0:
        #     Ad_headers.append(data)
        if i%2 != 0:
            COMPANY_DATA.append(data)
    print("Address Data -- ok")

    # COMPANY_DATA = [name, company_name]
    # for ele in MI_data:
    #     COMPANY_DATA.append(ele)
    # for ele in BI_data:
    #     COMPANY_DATA.append(ele)
    # for ele in Ad_data:
    #     COMPANY_DATA.append(ele)
    COMPANY_DATA_2D.append(COMPANY_DATA)
    print("------PARSING STEP DONE------")
    print(COMPANY_DATA)
    print("\n")
```
