## Scraping **Trading Codes** to Generate the urls of different companies

```python
source = requests.get("https://www.dsebd.org/company_listing.php").text
soup = BeautifulSoup(source, "lxml")
Trading_Name = []
for trading_name in soup.find_all('a', class_='abhead'):
    Trading_Name.append(trading_name['href'].split("=")[1])
print(Trading_Name)
```

## Automatically Generating the urls of the companies

```python
Company_URL =[]
for name in Trading_Name:
    company_url = 'https://www.dsebd.org/displayCompany.php?name='+name
    print(company_url)
    Company_URL.append(company_url)
```

## SQLite Database

```python
import sqlite3
# Connect/Create database
connection = sqlite3.connect('DSE.db')
# Create cursor object to create database table further
cursor = connection.cursor()
```

```python
#Create Table
cursor.execute('''CREATE TABLE IF NOT EXISTS Company
 ('Trading Code' TEXT NOT NULL,
 'Company Name' TEXT,
 'Date' TEXT NOT NULL, 
 'Time' TEXT NOT NULL,
 'Last Trading Price' TEXT,
 'Closing Price' TEXT,
 'Last Update' TEXT,
 "Day's Range" TEXT,
 'Change*' TEXT,
 "Day's Value (mn)" TEXT,
 'Change(%)' TEXT,
 "52 Weeks' Moving Range" TEXT,
 'Opening Price' TEXT,
 "Day's Volume (Nos.)" TEXT,
 'Adjusted Opening Price' TEXT,
 "Day's Trade (Nos.)" TEXT,
 "Yesterday's Closing Price" TEXT,
 'Market Capitalization (mn)' TEXT,
 'Authorized Capital (mn)' TEXT,
 'Debut Trading Date' TEXT,
 'Paid-up Capital (mn)' TEXT,
 'Type of Instrument' TEXT,
 'Face/par Value' INTEGER,
 'Market Lot' REAL,
 'Total No. of Outstanding Securities' TEXT,
 'Sector' TEXT,
 'Address' TEXT,
 'Contact Phone' TEXT,
 'Fax' TEXT,
 'E-mail' TEXT,
 'Web Address' TEXT,
 'Company Secretary Name' TEXT,
 'Cell No.' TEXT,
 'Telephone No.' TEXT,
 'Secretary Email' TEXT,
 PRIMARY KEY ('Trading Code', 'Date', 'Time'))''')
```

## Creating DATAFRAME from SQLite Database and Ecporting to CSV

```python
import pandas as pd

# Connect/Create database
connection = sqlite3.connect('DSE.db')
# Create cursor object to create database table further
cursor = connection.cursor()

sql_query  = pd.read_sql_query('''SELECT * FROM Company''', connection)

df_from_sql = pd.DataFrame(sql_query, columns = HEADER)
connection.close() # Closing connection

# Exporting dataframe to CSV
df_from_sql.to_csv('DSE Scraped Data_Taufiq.csv', index=False)
df_from_sql.to_excel("DSE Scraped Data_Taufiq.xlsx", index=False)

df_from_sql

# # Creating dataframe from above lists
# df = pd.DataFrame(COMPANY_DATA_2D, columns=HEADER)
```
